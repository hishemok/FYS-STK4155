Show that $$\frac{∂ \mathbf{b}^T \mathbf{a}}{\partial\mathbf{a}} = \mathbf{b}$$

$$\frac{\partial\mathbf{b}^T\mathbf{a}}{\partial\mathbf{a}} = \frac{\partial(b_0a_0 + b_1a_1 + ... + b_na_n)}{\partial\mathbf{a}} =\begin{bmatrix}
\frac{∂ (b_0a_0 + b_1a_1 + ... + b_na_n)}{∂ a_0} \\
\frac{∂ (b_0a_0 + b_1a_1 + ... + b_na_n)}{∂ a_1} \\
\vdots \\
\frac{∂ (b_0a_0 + b_1a_1 + ... + b_na_n)}{∂ a_n} \\
\end{bmatrix} =\begin{bmatrix}
b_0 \\
b_1 \\
\vdots \\
b_n
\end{bmatrix} = \mathbf{b}$$

Then show that: $$\frac{∂ \mathbf{a^T}\mathbf{A}\mathbf{a} }{∂ \mathbf{a}} = \mathbf{a}^T (\mathbf{A} + \mathbf{A}^T)$$

$$\mathbf{a}^T\mathbf{A}\mathbf{a} = [a_0,a_1, ... , a_n] 
\begin{bmatrix} 
A_{00} & A_{01} & ... & A_{0n}\\ 
A_{10} & A_{11} & ... & A_{1n}\\
\vdots & \vdots & \vdots & \vdots\\ 
A_{n0} & A_{n1} & ... & A_{nn}\\ 
\end{bmatrix} 
\begin{bmatrix}
a_0 \\
a_1 \\
\vdots \\
a_n
\end{bmatrix}$$
$$ =  [a_0,a_1, ... , a_n] 
\begin{bmatrix} 
\sum_{i} A_{0i}a_i\\ 
\sum_{i} A_{1i}a_i\\
\vdots \\ 
\sum_{i} A_{ni}a_i\\ 
\end{bmatrix}  = \sum_{ij}\left(a_jA_{ji}a_i \right) $$ 
$$\frac{∂ }{∂ a_k} \sum_{ij}\left(a_jA_{ji}a_i \right) = ∑_{j}A_{kj}a_j + ∑ _{i}A{ki}a_i = \mathbf{a}^T(\mathbf{A} + \mathbf{A}^T)$$

Lastly, show that: $$\frac{∂ (\mathbf{x} - \mathbf{As})^T(\mathbf{x}-\mathbf{As})}{∂ \mathbf{s}} = -2(\mathbf{x}-\mathbf{As})^T\mathbf{A}$$

To show that $$\frac{∂ (\mathbf{x} - \mathbf{As})^T(\mathbf{x}-\mathbf{As})}{∂ \mathbf{s}} = -2(\mathbf{x}-\mathbf{As})^T\mathbf{A}$$

Let's start by expanding the expression on the left-hand side:

$$\frac{∂ (\mathbf{x} - \mathbf{As})^T(\mathbf{x}-\mathbf{As})}{∂ \mathbf{s}} = \frac{∂ (\mathbf{x}^T\mathbf{x} - \mathbf{x}^T\mathbf{As} - (\mathbf{As})^T\mathbf{x} + (\mathbf{As})^T\mathbf{As})}{∂ \mathbf{s}}$$

1. $\frac{∂ \mathbf{x}^T\mathbf{x}}{∂ \mathbf{s}} = 0$ since $\mathbf{x}$ is not a function of $\mathbf{s}$. 
2. $\mathbf{x}^T\mathbf{As} =  (\mathbf{As})^T\mathbf{x}$, since both are scalars, so $$\frac{∂ - \mathbf{x}^T\mathbf{As} - (\mathbf{As})^T\mathbf{x}}{∂ \mathbf{s}} = -2 \frac{∂ \mathbf{x}^T\mathbf{As}}{∂\mathbf{s}} =  -2\mathbf{A}^T\mathbf{x}$$


3. $$\frac{∂ (\mathbf{As})^T\mathbf{As}}{∂ \mathbf{s}} = \frac{∂ \mathbf{s}^T\mathbf{A}^T\mathbf{As}}{∂ \mathbf{s}} = \sum_j  \frac{∂ s_j}{∂ s_k}\sum_i (A^TA)_{ij}s_i +  \sum_j s_j \sum_i (A^TA)_{ij}\frac{∂ s_i}{∂ s_k}$$
Since $(A^TA)$ is symmetric, we can combine the two sums:
$$\frac{∂ (\mathbf{As})^T\mathbf{As}}{∂ \mathbf{s}} = 2\sum_i (A^TA)_{ki}s_i = 2(A^TA)_{k:}\mathbf{s} = 2\mathbf{A}^T\mathbf{As}$$

4. Now we can put it all together:

$$\frac{∂ (\mathbf{x} - \mathbf{As})^T(\mathbf{x}-\mathbf{As})}{∂ \mathbf{s}} = -2\mathbf{A}^T\mathbf{x} + 2\mathbf{A}^T\mathbf{As} = -2(\mathbf{x}-\mathbf{As})^T\mathbf{A}$$

<h2> Exercise 2: making your own data and exploring scikit-learn</h2>

```python
#imports
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
```

```python
np.random.seed(7)

n = 100
x = np.random.rand(n,1)
x = np.sort(x,axis=0)
y = 1.0 + 3*x**2 + 0.1*np.random.randn(n,1) #noise added to true function
yt = 1.0 + 3*x**2 #true function
```

```python
#first order for fun
X = np.ones((n,2))
X[:,1] = x[:,0]

beta = np.linalg.inv(X.T@X)@X.T@y
ytilde = X@beta
```

```python
plt.plot(x,y,'o',label='Datapoints',markersize=3)
plt.plot(x,ytilde,label='Prediction')
plt.plot(x,yt,label='True function')
plt.legend()
plt.show()
```

```python
#second order, with sklearn and without
poly2 = PolynomialFeatures(degree=2)
X = poly2.fit_transform(x)
linreg2 = LinearRegression()
linreg2.fit(X,y)
y_pred2 = linreg2.predict(X)

X = np.ones((n,3))
X[:,1] = x[:,0]
X[:,2] = x[:,0]**2

beta = np.linalg.inv(X.T@X)@X.T@y
ytilde2 = X@beta
```

```python
plt.plot(x,y,'o',label='Datapoints',markersize=3)
plt.plot(x,y_pred2,label='Prediction scikit')
plt.plot(x,ytilde2,label='Prediction manual')
plt.plot(x,yt,label='True function')
plt.legend()
plt.show()
```

```python
#errors on the sklearn model
mse = mean_squared_error(yt,ytilde)
msem2 = mean_squared_error(yt,ytilde2)
mses2 = mean_squared_error(yt,y_pred2)
print(f'MSE linear: {mse:.5f}')
print(f'MSE poly manual:{msem2:.5f}')
print(f'MSE poly sklearn:{mses2:.5f}')


r2 = r2_score(yt,ytilde)
rm22 = r2_score(yt,ytilde2)
rs22 = r2_score(yt,y_pred2)
print(f'R2 linear:{r2:.5f}')
print(f'R2 poly manual:{rm22:.5f}')
print(f'R2 poly sklearn:{rs22:.5f}')
```

<h2> Exercise 3: Split data in test and training data</h2>

<h3>Design matrix and split data</h3>

```python
#import
from sklearn.model_selection import train_test_split
```

```python
n = 100 

# Generate data
x = np.linspace(-3, 3, n).reshape(-1, 1)
y = np.exp(-x**2) + 1.5 * np.exp(-(x-2)**2) + np.random.normal(0, 0.1, x.shape)
```

```python

X = np.ones((n,6))
X[:,1] = x.flatten()
X[:,2] = x.flatten()**2
X[:,3] = x.flatten()**3
X[:,4] = x.flatten()**4
X[:,5] = x.flatten()**5

#split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) #80% training, 20% test
```

<h3>Fit the model </h3>

```python
#sort data
sort = np.argsort(X_train[:,1])
X_train = X_train[sort]
y_train = y_train[sort]

sort = np.argsort(X_test[:,1])
X_test = X_test[sort]
y_test = y_test[sort]

# matrix inversion to find beta
beta = np.linalg.inv(X.T @ X) @ X.T @ y
# and then make the prediction
ytilde = X_train @ beta
ypred = X_test @ beta

#plot
plt.plot(x, y,'o', label='True function',markersize=3)
plt.plot(X_train[:,1], ytilde, label='Model prediction (train)')
plt.plot(X_test[:,1], ypred, label='Model prediction (test)')
plt.legend()
plt.show()

```

```python
mse_train = mean_squared_error(y_train, ytilde)
mse_test = mean_squared_error(y_test, ypred)

print(f'MSE for training data: {mse_train:.2e}')
print(f'MSE for test data: {mse_test:.2e}')
```
