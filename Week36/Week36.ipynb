<h3>a) Expression for Ridge regression</h3>

Show that the optimal parameters $$\mathbf{\hat{\beta}}_{Ridge} = (\mathbf{X}^T\mathbf{X} + λ \mathbf{I})^{-1} \mathbf{X}^T\mathbf{y}$$
with $\mathbf{I}$ being a $p × p$ identity matrix.

The ordinary least squares result is $$\mathbf{\hat{\beta}}_{OLS} = (\mathbf{X}^T\mathbf{X})^{-1} \mathbf{X}^T\mathbf{y}$$

To show that the optimal parameters for Ridge regression are given by:

$$\mathbf{\hat{\beta}}_{Ridge} = (\mathbf{X}^T\mathbf{X} + λ \mathbf{I})^{-1} \mathbf{X}^T\mathbf{y}$$

we need to derive this from the Ridge regression cost function. 

The cost function for Ridge regression is given by: $$ C(\mathbf{\beta}) = (\mathbf{y} - \mathbf{X}\mathbf{\beta})^T (\mathbf{y} - \mathbf{X}\mathbf{\beta}) + λ \mathbf{\beta}^T \mathbf{\beta} $$

Expand the Cost Function: $$ C(\mathbf{\beta}) = (\mathbf{y}^T - \mathbf{\beta}^T \mathbf{X}^T)(\mathbf{y} - \mathbf{X}\mathbf{\beta}) + λ \mathbf{\beta}^T \mathbf{\beta} $$ $$ C(\mathbf{\beta}) = \mathbf{y}^T \mathbf{y} - \mathbf{y}^T \mathbf{X} \mathbf{\beta} - \mathbf{\beta}^T \mathbf{X}^T \mathbf{y} + \mathbf{\beta}^T \mathbf{X}^T \mathbf{X} \mathbf{\beta} + λ \mathbf{\beta}^T \mathbf{\beta} $$

then combine the terms: $$ C(\mathbf{\beta}) = \mathbf{y}^T \mathbf{y} - 2 \mathbf{y}^T \mathbf{X} \mathbf{\beta} + \mathbf{\beta}^T (\mathbf{X}^T \mathbf{X} + λ \mathbf{I}) \mathbf{\beta} $$

Take the Gradient with Respect to $\mathbf{\beta}$ and set equal to zero for optimization: $$ \frac{\partial C(\mathbf{\beta})}{\partial \mathbf{\beta}} = -2 \mathbf{X}^T \mathbf{y} + 2 (\mathbf{X}^T \mathbf{X} + λ \mathbf{I}) \mathbf{\beta}  = 0$$

$$ -2 \mathbf{X}^T \mathbf{y} + 2 (\mathbf{X}^T \mathbf{X} + λ \mathbf{I}) \mathbf{\beta} = 0 $$ 
$$ (\mathbf{X}^T \mathbf{X} + λ \mathbf{I}) \mathbf{\beta} = \mathbf{X}^T \mathbf{y} $$

Solve for $(\mathbf{\beta})$: $$ \mathbf{\beta} = (\mathbf{X}^T \mathbf{X} + λ \mathbf{I})^{-1} \mathbf{X}^T \mathbf{y} $$

Thus, we have shown that the optimal parameters for Ridge regression are:

$$\mathbf{\hat{\beta}}_{Ridge} = (\mathbf{X}^T\mathbf{X} + λ \mathbf{I})^{-1} \mathbf{X}^T\mathbf{y}$$

<h3>b) The singular value decomposition</h3>

<h3>OLS</h3>

The singular value decomposition of an $n × p$ matrix $\mathbf{X}$:
$$\mathbf{X} = \mathbf{UΣV}^T$$
$\mathbf{U}$ and $\mathbf{V}$ are orthogonal matrices of dimensions $n × n$ and $p × p$, and $\mathbf{\Sigma}$ is an $n× p$ matrix which contains the singular values only.

First we show that you can write the OLS solutions in terms of the eigenvectors (the columns) of the orthogonal matrix $\mathbf{U}$ as
$$\mathbf{\tilde{y}}_{OLS} =\mathbf{X\beta} = \sum_{j=0}^{p-1} \mathbf{u_j} \mathbf{u_j}^T\mathbf{y}$$

The OLS solution for $\mathbf{\beta}$ is given by: $$\mathbf{\hat{\beta}} = (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T \mathbf{y}$$

First we can find $\mathbf{X^TX}$: $$\mathbf{X}^T \mathbf{X} = (\mathbf{UΣV}^T)^T (\mathbf{UΣV}^T) = \mathbf{VΣ}^T \mathbf{U}^T \mathbf{UΣV^T} = \mathbf{VΣ}^T \mathbf{ΣV^T}$$ 
Since $\mathbf{U}$ is orthogonal, $\mathbf{U}^T \mathbf{U} = \mathbf{I}$.

Inverse of $\mathbf{X}^T \mathbf{X}$: $$ (\mathbf{X}^T \mathbf{X})^{-1} = (\mathbf{VΣ}^T \mathbf{ΣV^T})^{-1} = \mathbf{V}^{-T} (\mathbf{Σ}^T \mathbf{Σ})^{-1} \mathbf{V}^{-1} =  \mathbf{V} (\mathbf{Σ}^T \mathbf{Σ})^{-1} \mathbf{V}^{T}$$ Since $\mathbf{V}$ is orthogonal, $\mathbf{V}^{-1} = \mathbf{V}^T$.

Substitute Back into OLS Solution: $$\mathbf{\hat{\beta}} = \mathbf{V} (\mathbf{Σ}^T \mathbf{Σ})^{-1} \mathbf{V}^{T} \mathbf{X}^T \mathbf{y}$$ 
$$\mathbf{\hat{\beta}} = \mathbf{V} (\mathbf{Σ}^T \mathbf{Σ})^{-1} \mathbf{V}^{T} \mathbf{V\Sigma^TU^T} \mathbf{y}$$ 
$$\mathbf{\hat{\beta}} = \mathbf{V} (\mathbf{Σ}^T \mathbf{Σ})^{-1} \mathbf{Σ}^T \mathbf{U}^T \mathbf{y}$$

Compute $\mathbf{\tilde{y}}_{OLS}$: $$\mathbf{\tilde{y}}_{OLS} = \mathbf{X\hat{\beta}} = \mathbf{UΣV}^T \mathbf{V} (\mathbf{Σ}^T \mathbf{Σ})^{-1} \mathbf{Σ}^T \mathbf{U}^T \mathbf{y}$$ 
$$\mathbf{\tilde{y}}_{OLS} = \mathbf{UΣ} (\mathbf{Σ}^T \mathbf{Σ})^{-1} \mathbf{Σ}^T \mathbf{U}^T \mathbf{y}$$ 
And since $(\mathbf{\Sigma^T\Sigma}) = diag(\sigma_0^2,\sigma_1^2,...,σ_i^2)$ 

and $(\mathbf{\Sigma^T\Sigma})^{-1} = diag(1/\sigma_0^2,1/\sigma_1^2,...,1/σ_i^2)$ then $\mathbf{\Sigma}(\mathbf{Σ}^T \mathbf{Σ})^{-1} \mathbf{Σ}^T= \mathbf{I}$, so:

$$\mathbf{\tilde{y}}_{OLS} = \mathbf{U} \mathbf{U}^T \mathbf{y}$$

Express in Terms of Eigenvectors: Since $\mathbf{U}$ is orthogonal, it can be written as a sum of its eigenvectors: $$\mathbf{U} \mathbf{U}^T = \sum_{j=0}^{p-1} \mathbf{u_j} \mathbf{u_j}^T$$

Final Expression: $$\mathbf{\tilde{y}}_{OLS} = \sum_{j=0}^{p-1} \mathbf{u_j} \mathbf{u_j}^T \mathbf{y}$$


<h3>Ridge</h3>

For Ridge regression, show that the corresponding equation is 
$$\mathbf{\tilde{y}}_{Ridge} = \mathbf{X\beta}_{Ridge} = \mathbf{UΣ V}^T(\mathbf{VΣ^2 V^T } + λ \mathbf{I})^{-1} (\mathbf{U Σ V^T})^T\mathbf{y} = \sum_{j=1}^{p-1}\mathbf{u_ju_j^T} \frac{σ_j^2 }{\sigma_j^2 +\lambda} \mathbf{y}$$

First we start with:
$$\mathbf{UΣ V}^T(\mathbf{VΣ^2 V^T } + λ \mathbf{I})^{-1} (\mathbf{U Σ V^T})^T\mathbf{y}$$
and re-write to: $$\mathbf{UΣ V}^TV(\mathbf{Σ^2  } + λ \mathbf{I})^{-1} V^T(\mathbf{U Σ V^T})^T\mathbf{y}$$

Then we can further simplify:
$$\mathbf{U\Sigma(\Sigma^2+λ I)^{-1}\Sigma^TU^Ty}$$
Now we can take a look at $\mathbf{\Sigma}(\mathbf{\Sigma}^2 +λ \mathbf{I})^{-1} \Sigma^T$ element by element,
$$\mathbf{\Sigma}(\mathbf{\Sigma}^2 +λ \mathbf{I})^{-1} \Sigma^T = (\frac{\sigma_0^2}{\sigma_0^2 + λ} , \frac{\sigma_1^2}{\sigma_1^2 + λ}, \frac{\sigma_i^2}{\sigma_i^2 + λ})$$
Final expression:
$$\mathbf{\tilde{y}}_{Ridge} = \mathbf{UΣ V}^TV(\mathbf{Σ^2  } + λ \mathbf{I})^{-1} V^T(\mathbf{U Σ V^T})^T\mathbf{y} = \sum_{j=1}^{p-1}\mathbf{u_j}\mathbf{u_j}^T\frac{\sigma_j^2}{\sigma_j^2 + λ} \mathbf{y}$$

<h3>Exercise 2: Adding Ridge Regression</h3>

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error as mse


#define a function
np.random.seed()
n = 100
# Make data set.
x = np.linspace(-3, 3, n).reshape(-1, 1)
y = np.exp(-x**2) + 1.5 * np.exp(-(x-2)**2) + np.random.normal(0, 0.1, x.shape)
```

```python
X = np.ones((n,6))
X[:,1] = x[:,0]
X[:,2] = x[:,0]**2
X[:,3] = x[:,0]**3
X[:,4] = x[:,0]**4
X[:,5] = x[:,0]**5


def scaling(X,y):
    xmean = np.mean(X, axis=0)
    xstd = np.std(X, axis=0)
    xstd[xstd == 0] = 1 #avoid division by zero
    ymean = np.mean(y)
    ystd = np.std(y)
    X_scaled = (X - xmean)/xstd
    y_scaled = (y - ymean)/ystd
    X_scaled[:,0] = 1
    return X_scaled, y_scaled, xmean, xstd, ymean, ystd

X_scaled, y_scaled, xmean, xstd, ymean, ystd = scaling(X,y)
print(X_scaled.shape)

```

```python
#test train split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.2)
sort = np.argsort(X_train[:,1])
X_train = X_train[sort]
y_train = y_train[sort]

sort = np.argsort(X_test[:,1])
X_test = X_test[sort]
y_test = y_test[sort]
```

```python
#calculate the beta values for OLS and Ridge
lam = [0.0001, 0.001, 0.01, 0.1, 1]
betaOLS = []
betaRidge = []

for i in lam:
    ols  = np.linalg.inv(X_train.T @ X_train) @ X_train.T @ y_train
    ridge = np.linalg.inv(X_train.T @ X_train + i*np.eye(6)) @ X_train.T @ y_train

    betaOLS.append(ols)
    betaRidge.append(ridge)
```

```python
for i in range(len(lam)):
    ytildeOLS = X_train @ betaOLS[i]
    ytildeRidge = X_train @ betaRidge[i]
    
    ypredictOLS = X_test @ betaOLS[i]
    ypredictRidge = X_test @ betaRidge[i]

    MSEOLStest = mse(y_test, ypredictOLS)
    MSERidgetest = mse(y_test, ypredictRidge)
    MSEOLStrain = mse(y_train, ytildeOLS)
    MSERidgetrain = mse(y_train, ytildeRidge)

    print(f'MSE for OLS with   lambda  = None   is |test: {MSEOLStest:.4f}|  train: {MSEOLStrain:.4f}|')
    print(f'MSE for Ridge with lambda  = {lam[i]:.4f} is |test: {MSERidgetest:.4f}|  train: {MSERidgetrain:.4f}|')
    print('------------------------------------')

# #unscale
X_train = X_train*xstd + xmean
X_test = X_test*xstd + xmean
ytildeOLS = ytildeOLS*ystd + ymean
ytildeRidge = ytildeRidge*ystd + ymean
ypredictOLS = ypredictOLS*ystd + ymean
ypredictRidge = ypredictRidge*ystd + ymean


#plot the data
plt.plot(x, y, 'ro',markersize = 3)
plt.plot(X_train[:,1], ytildeOLS, label='Train')
plt.plot(X_test[:,1], ypredictOLS, label='Predicted')
plt.title('OLS')
plt.legend()
plt.show()

#plot the data
plt.plot(x, y, 'ro',markersize = 3)
plt.plot(X_train[:,1], ytildeRidge, label='Train')
plt.plot(X_test[:,1], ypredictRidge, label='Predicted')
plt.title('Ridge')
plt.legend()
plt.show()
```
